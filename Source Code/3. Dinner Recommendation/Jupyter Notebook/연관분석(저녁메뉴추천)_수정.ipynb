```python
dt_menus = []
unique_menus = []

with open(os.path.join(data_path, "dt_menus.txt"), 'r', encoding="euc-kr") as f:
    for line in f:
      line_menus = line.replace("\n","").split(",")
      dt_menus.append(line_menus)
      unique_menus.extend(line_menus)

values, counts = np.unique(unique_menus, return_counts=True)

# 내림차순으로 정렬된 인덱스 계산
sorted_indexes = np.argsort(counts)[::-1]

# 정렬된 값과 해당 값의 빈도 출력
sorted_values = values[sorted_indexes]
sorted_counts = counts[sorted_indexes]

for value, count in zip(sorted_values, sorted_counts):
    print(f'{value}: {count}')

unique_menus = list(set(unique_menus))
len(dt_menus)
```

```python
# 아이템의 빈도를 기반으로 가중치를 할당하는 함수
def assign_weights(itemsets):
    item_frequencies = {}
    for transaction in itemsets:
        for item in transaction:
            if item in item_frequencies:
                item_frequencies[item] += 1
            else:
                item_frequencies[item] = 1

    weights = {}
    for item, frequency in item_frequencies.items():
        weight = 1 / frequency
        weights[item] = weight

    return weights

# 가중치 할당
weights = assign_weights(dt_menus)

# 가중치가 적용된 아이템셋 데이터
weighted_itemsets = []
for transaction in dt_menus:
    weighted_transaction = [item for item in transaction if item in weights]
    weighted_itemsets.append(weighted_transaction)
```

```python
te = TransactionEncoder()

te_ary = te.fit_transform(weighted_itemsets)

df = pd.DataFrame(te_ary, columns=te.columns_)

te = None
te_ary = None
```

```python
# 빈발 아이템셋 추출
frequent_itemsets = fpgrowth(df, min_support=1e-5, use_colnames=True)
# frequent_itemsets = apriori(df, min_support=1e-3, max_len=int(1e+3), use_colnames=True)
# df = None
# 연관 규칙 추출
rules = association_rules(frequent_itemsets, metric="conviction", min_threshold=1e-5)
# frequent_itemsets = None
```

```python
# 반찬 추천 함수 정의
def recommend_side_dish_confidence(input_dish, top_n=10):
  str_menu = " ".join(unique_menus)
  find_matches = re.findall(f"[가-힣]*{input_dish}[가-힣]*", str_menu)

  for menu_idx, menu in enumerate(find_matches):
    print("="*50)

    if menu not in list(rules['antecedents'].str.join(',')):
      print(f"입력한 메뉴({menu})가 학습 데이터에 없습니다.")
      continue

    print(f"입력한 메뉴({menu})의 반찬은 다음과 같습니다.")
    recommended_dishes = rules[rules['antecedents'].str.join(',') == menu].sort_values('confidence', ascending=False)[:top_n]
    print(recommended_dishes['consequents'].str.join(',').tolist())

# 반찬 추천 함수 정의
def recommend_side_dish_conviction(input_dish, top_n=10):
  str_menu = " ".join(unique_menus)
  find_matches = re.findall(f"[가-힣]*{input_dish}[가-힣]*", str_menu)

  for menu_idx, menu in enumerate(find_matches):
    print("="*50)

    if menu not in list(rules['antecedents'].str.join(',')):
      print(f"입력한 메뉴({menu})가 학습 데이터에 없습니다.")
      continue

    print(f"입력한 메뉴({menu})의 반찬은 다음과 같습니다.")
    recommended_dishes = rules[rules['antecedents'].str.join(',') == menu].sort_values('conviction', ascending=False)[:top_n]
    print(recommended_dishes['consequents'].str.join(',').tolist())
```
